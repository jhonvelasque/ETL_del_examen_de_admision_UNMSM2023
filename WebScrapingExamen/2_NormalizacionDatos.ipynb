```python
#Inputacion de el orde de merito 
df_resultados_imputado=pd.DataFrame()
for modalidad in df_resultados['IdModalidad'].unique():
    for carrera in df_resultados['IdCarrera'].unique():
        df_auxiliar=df_resultados[(df_resultados['IdModalidad']==modalidad) & (df_resultados['IdCarrera']==carrera)].sort_values(by=['PUNTAJE FINAL','APELLIDOS Y NOMBRES'],ascending=False).reset_index( )
        df_auxiliar['Merito']=[i+1 for i in range(len(df_auxiliar))]
        df_resultados_imputado = pd.concat([df_resultados_imputado, df_auxiliar], ignore_index=True)
```
